```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction import DictVectorizer
from scipy.sparse import hstack
load_train_data = pd.read_csv('salary_train.csv')
```

```python
load_train_data
```

```python
load_train_data['FullDescription'] = [i.lower() for i in load_train_data.iloc[:,0]]
load_train_data['FullDescription'] = load_train_data['FullDescription'].replace('[^a-z0-9]', ' ', regex = True)
```

```python
load_train_data
```

```python
train_fulldescript_vectorized = load_train_data['FullDescription']
vectorizer = TfidfVectorizer(min_df=5)
train_fulldescript_vectorized = vectorizer.fit_transform(train_fulldescript_vectorized)
load_train_data['LocationNormalized'].fillna('nan', inplace=True)
load_train_data['ContractTime'].fillna('nan', inplace=True)
enc = DictVectorizer()
train_categ_vectorized = enc.fit_transform(load_train_data[['LocationNormalized', 'ContractTime']].to_dict('records'))
X_train = hstack([train_fulldescript_vectorized, train_categ_vectorized])
y_train = load_train_data['SalaryNormalized']
```

```python
load_test_data = pd.read_csv('salary_test_mini.csv')
load_test_data
```

```python
load_test_data['FullDescription'] = [i.lower() for i in load_test_data.iloc[:,0]]
load_test_data['FullDescription'] = load_test_data['FullDescription'].replace('[^a-z0-9]', ' ', regex = True)

test_fulldescript_vectorized = load_test_data['FullDescription']
test_fulldescript_vectorized = vectorizer.transform(test_fulldescript_vectorized)
load_test_data['LocationNormalized'].fillna('nan', inplace=True)
load_test_data['ContractTime'].fillna('nan', inplace=True)
test_categ_vectorized = enc.transform(load_test_data[['LocationNormalized', 'ContractTime']].to_dict('records'))
X_test = hstack([test_fulldescript_vectorized, test_categ_vectorized])
```

```python
from sklearn.linear_model import Ridge
model = Ridge(alpha=1,random_state=241).fit(X_train, y_train)
res = model.predict(X_test)
```
