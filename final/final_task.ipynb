```python
import pandas
import math
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import warnings
from sklearn.model_selection import KFold
warnings.filterwarnings('ignore')
features = pandas.read_csv('./features.csv', index_col='match_id')
print(features.shape)
features.head()
```

## GradientBoostingClassifier

```python
for i in features.columns:
    print(i, " - ", sum(features[i] != features[i]))
```

```python
y = features['radiant_win']
y
```

```python
X = features.copy()
del X['radiant_win']
X = X.fillna(0)
```

```python
gb = GradientBoostingClassifier(n_estimators=30, random_state=241)
from sklearn.model_selection import KFold
kf = KFold(random_state = 42, shuffle = True)
```

```python
from datetime import datetime
start_time = datetime.now()
quality = cross_val_score(gb, X, y, cv=kf, scoring='f1')
print(datetime.now() - start_time)
np.mean(quality)
```

```python
gb = GradientBoostingClassifier(n_estimators=40, random_state=241)
start_time = datetime.now()
quality = cross_val_score(gb, X, y, cv=kf, scoring='f1', n_jobs = -1)
print(datetime.now() - start_time)
np.mean(quality)
```

## Logistic

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
start_time = datetime.now()
quality = cross_val_score(lr, X, y, cv=kf, scoring='f1')
print(datetime.now() - start_time)
np.mean(quality)
```

```python
import re
X_without_categories = X.copy()
all_cat = [r'r1', r'r2', r'r3', r'r4', r'r5', r'd1', r'd2', r'd3', r'd4', r'd5']
for j in all_cat:
    for i in X.columns:
        if re.match(j, i): del X_without_categories[i]
X_without_categories
```

```python
start_time = datetime.now()
quality = cross_val_score(lr, X_without_categories, y, cv=kf, scoring='f1')
print(datetime.now() - start_time)
np.mean(quality)
```
