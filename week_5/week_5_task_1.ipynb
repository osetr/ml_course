```python
import pandas as pd
import numpy as np
data = pd.read_csv('abalone.csv')
data.loc[data['Sex'] == 'M',('Sex')] = 1
data.loc[data['Sex'] == 'F',('Sex')] = -1
data.loc[data['Sex'] == 'I',('Sex')] = 0
data
```

```python
X = data[data.columns[:8]]
X
```

```python
y = data[data.columns[8:]]
y
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import r2_score
import warnings
warnings.filterwarnings('ignore')
kf = KFold(random_state = 1, shuffle = True)
kf.get_n_splits(X)
regressor = RandomForestRegressor(random_state=1, n_estimators=21)
quality = cross_val_score(regressor, X, y, cv=kf, scoring='r2')
quality.mean()
```
